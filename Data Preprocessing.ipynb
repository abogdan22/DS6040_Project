```python
#remove last column
GVA_online = GVA_online[['Incident ID','Incident Date','State','City Or County','Address','# Killed','# Injured']]
```

```python
#lower case column names
GVA_online = GVA_online.rename(columns = str.lower)
```

```python
#remove spaces from column names
GVA_online.columns = GVA_online.columns.str.replace(" ","")
```

```python
#GVA_online[GVA_online['incidentdate'] == 'June 29, 2022']
GVA_online
```

```python
#Subset to 2020+
years = ('20', '21', '22') #create a tuple (because .str.endswith accepts a tuple of suffixes, not a list)
```

```python
GVA_recent = GVA_online[GVA_online['incidentdate'].str.endswith(years)]
```

```python
GVA_recent['total_involved'] = GVA_recent['#killed'] + GVA_recent['#injured']
```

```python
accidents = pd.DataFrame(GVA_recent.groupby('state')['total_involved'].sum()) #might not need to create a dataframe
```

```python
# Resetting index so I can rename the columns
accidents = accidents.reset_index()
```
